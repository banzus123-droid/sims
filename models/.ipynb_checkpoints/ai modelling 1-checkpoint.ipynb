```python
import pandas as pd
import numpy as np
import re
import string
import nltk
import joblib
import os
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score

# --- PHASE 1: HIGH-FIDELITY RESOURCES ---
# We use Lemmatizer for semantic accuracy and keep emojis for emotional context
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Dictionary for Contraction Expansion
CONTRACTIONS = {
    "can't": "cannot", "won't": "will not", "n't": " not",
    "i'm": "i am", "it's": "it is", "i've": "i have",
    "don't": "do not", "doesn't": "does not", "didn't": "did not",
    "i'll": "i will", "you're": "you are", "they're": "they are"
}

def clean_and_preprocess(text):
    """
    High-Fidelity Preprocessing Station:
    1. Contraction Expansion
    2. URL/Handle Noise Removal (Preserving Emojis)
    3. Tokenization
    4. Lemmatization (Semantic root extraction)
    """
    if not isinstance(text, str): return ""
    
    # 1. Expand Contractions
    text = text.lower()
    for contraction, expansion in CONTRACTIONS.items():
        text = text.replace(contraction, expansion)
    
    # 2. Advanced Noise Removal
    # This regex removes URLs and @handles but PRESERVES Unicode (Emojis)
    text = re.sub(r'http\S+|www\S+|\@\w+', '', text)
    # Remove numbers but keep emotional punctuation if necessary
    text = re.sub(r'\d+', '', text)
    
    # 3. Tokenization
    tokens = word_tokenize(text)
    
    # 4. Lemmatization & Stopword Filtering
    # We keep characters that are either words (non-punctuation) or Emojis (non-ASCII)
    processed = []
    for w in tokens:
        if w not in stop_words:
            # Check if it's an emoji (non-ascii) or a word (alphanumeric)
            if any(ord(char) > 127 for char in w) or w.isalnum():
                processed.append(lemmatizer.lemmatize(w))
    
    return " ".join(processed)

def map_risk_labels(status):
    """Maps various dataset labels into 3 standard categories: 0=Low, 1=Moderate, 2=High"""
    status = str(status).lower()
    # High Risk Markers
    if any(x in status for x in ['suicide', 'suicidal', 'high']): return 2
    # Moderate Risk Markers
    if any(x in status for x in ['anxiety', 'stress', 'depression', 'moderate', 'medium']): return 1
    # Low Risk
    return 0

# --- PHASE 2: DATA INGESTION ---
def load_all_datasets():
    files = [
        'Suicide_Ideation_Dataset(Twitter-based).csv',
        'dataset_part1.csv', 
        'dataset_part2.csv',
        'mental_health_combined_test.csv',
        'mental_heath_feature_engineered.csv',
        'mental_heath_unbanlanced.csv'
    ]
    
    frames = []
    for f in files:
        if os.path.exists(f):
            df = pd.read_csv(f)
            # Identify columns dynamically
            text_col = 'text' if 'text' in df.columns else 'tweet' if 'tweet' in df.columns else 'Tweet'
            label_col = 'label' if 'label' in df.columns else 'status' if 'status' in df.columns else 'class' if 'class' in df.columns else 'Suicide'
            
            if text_col and label_col in df.columns:
                df = df[[text_col, label_col]].rename(columns={text_col: 'text', label_col: 'raw_label'})
                df['label'] = df['raw_label'].apply(map_risk_labels)
                frames.append(df[['text', 'label']])
    
    if not frames:
        raise FileNotFoundError("No datasets found in directory.")
        
    combined_df = pd.concat(frames, ignore_index=True)
    print(f"[LOG] Total samples combined: {len(combined_df)}")
    
    print("[LOG] Running Preprocessing (Lemmatization + Emoji Support)...")
    combined_df['preprocessed_text'] = combined_df['text'].apply(clean_and_preprocess)
    
    # Save a sample to show the lecturer the "High-Fidelity" cleaning
    combined_df[['text', 'preprocessed_text', 'label']].head(200).to_csv('preprocessed_sample_hf.csv', index=False)
    
    return combined_df

# --- PHASE 3: ADVANCED ML TRAINING ---
def train_high_fidelity_model():
    df = load_all_datasets()
    
    # Filter empty rows
    df = df[df['preprocessed_text'] != ""]
    
    # Stratified split to handle imbalanced mental health classes
    X_train, X_test, y_train, y_test = train_test_split(
        df['preprocessed_text'], df['label'], test_size=0.2, random_state=42, stratify=df['label']
    )

    print("[LOG] Initializing High-Fidelity Random Forest Pipeline...")
    
    # Using Trigrams (1,3) to capture phrases like "i want to" or "help me please"
    # Random Forest is used for higher accuracy in non-linear text patterns
    sims_pipeline = Pipeline([
        ('tfidf', TfidfVectorizer(max_features=8000, ngram_range=(1, 3), sublinear_tf=True)),
        ('classifier', RandomForestClassifier(n_estimators=100, class_weight='balanced', n_jobs=-1))
    ])

    print("[LOG] Executing Training... (This may take several minutes due to dataset size)")
    sims_pipeline.fit(X_train, y_train)

    # --- EVALUATION ---
    predictions = sims_pipeline.predict(X_test)
    print("\n" + "="*40)
    print("HIGH-FIDELITY MODEL PERFORMANCE")
    print("="*40)
    print(f"Accuracy: {accuracy_score(y_test, predictions):.2%}")
    print(classification_report(y_test, predictions, target_names=['Low Risk', 'Moderate Risk', 'High Risk']))

    # --- SAVING ARTIFACTS ---
    print("\n[LOG] Saving artifacts for Dashboard...")
    joblib.dump(sims_pipeline, 'sims_pipeline.pkl')
    joblib.dump(sims_pipeline.named_steps['tfidf'], 'sims_tfidf_vectorizer.pkl')
    joblib.dump(sims_pipeline.named_steps['classifier'], 'sims_classifier.pkl')
    
    print("SUCCESS: High-Fidelity Engine is ready.")

if __name__ == "__main__":
    train_high_fidelity_model()
```
